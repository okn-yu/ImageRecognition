```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

from matplotlib import pyplot as plt
from skimage.io import imread, imshow
from skimage.feature import blob_dog, blob_log, blob_doh
```

# 画像の読み込み

```python
img =  cv2.imread('./sample.jpg', 0)[200:, :]
plt.imshow(img, cmap='gray')
plt.show()

_, mask_img = cv2.threshold(img, 0, 255, cv2.THRESH_OTSU)
mask_img = cv2.bitwise_not(mask_img)
plt.imshow(mask_img, cmap='gray')
plt.show()
```

# LoGフィルタ

```python
blobs = blob_log(mask_img, min_sigma=4)

fig, ax = plt.subplots(figsize=(6,6))
ax.imshow(img)

for blob in blobs:
    y, x, area = blob
    ax.add_patch(plt.Circle((x, y), area*np.sqrt(2), color='r', fill=False))
```

# DoGフィルタ

```python
blobs = blob_dog(mask_img, min_sigma=10)

fig, ax = plt.subplots(figsize=(6,6))
ax.imshow(img)
for blob in blobs:
    y, x, area = blob
    ax.add_patch(plt.Circle((x, y), area*np.sqrt(2), color='r', fill=False))
```

# DoH

```python
blobs = blob_doh(mask_img, min_sigma=4)
fig, ax = plt.subplots(figsize=(6,6))
ax.imshow(img)
for blob in blobs:
    y, x, area = blob
    ax.add_patch(plt.Circle((x, y), area*np.sqrt(2), color='r', fill=False))
```
